```python
import numpy as np

def valid_sudoku(board):
  #Check Board Shape
  if board.shape != (9, 9):
    return False

  #Check Values are 1-9 Only
  if not np.all((board >= 1) & (board <= 9)):
    return False

  #Check Rows
  for row in board:
    if len(np.unique(row)) != 9:
      return False

  #Check Columns
  for column in board.T:
    if len(np.unique(column)) != 9:
      return False

  #3x3 Sub Grids Check
  for i in range(0, 9, 3):
    for j in range(0, 9, 3):
      sub_grid = board[i:i+3, j:j+3].ravel()
      if len(np.unique(sub_grid)) != 9:
        return False

  return True

board = np.array([
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],
    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],
    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]
])

if valid_sudoku(board):
  print("Valid Sudoku")
else:
  print("Invalid Sudoku")
```
