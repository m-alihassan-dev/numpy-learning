```python
import numpy as np
```

Arithmetic Operations

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])

print("Sum:", arr1 + arr2)
print("Subtraction:", arr1 - arr2)
print("Multiplcation:", arr1 * arr2)
print("Division:", arr1 / arr2)
print("Floor Division:", arr1 // arr2)
print("Modulus:", arr1 % arr2)
print("Power:", arr1 ** arr2)
```

```python
arr1 = np.array([[1, 2, 3],
                 [4, 5, 6]])
arr2 = np.array([[7, 8, 9],
                 [10, 11, 12]])

print("Sum:", arr1 + arr2)
print("Subtraction:",arr1 - arr2)
print("Multiplication:", arr1 * arr2)
print("Division",arr2 / arr1)
```

Broadcasting

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr * 2)
```

Comparision Operators

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(arr % 2 == 0)
```

Matrix Multiplication

```python
arr1 = np.array([[1, 2],
                 [4, 5]])
arr2 = np.array([[7, 8],
                 [10, 11]])

print(arr1 @ arr2) # Or np.dot(arr1, arr2)
```
