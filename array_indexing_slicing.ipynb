```python
import numpy as np

arr = np.arange(1, 10)
print(arr)
```

Indexing of Vectors

```python
print(arr[5])
```

```python
print(arr[-1])
```

Slicing of Vectors

```python
print(arr[3:9])
```

```python
print(arr[:5])
```

```python
print(arr[5:])
```

```python
print(arr[::-1])
```

Indexing of Matrix

```python
arr = np.arange(1, 31).reshape(5, 6)
print(arr)
```

```python
arr[1, 1]
```

```python
arr[4, 5]
```

```python
arr[1] #Row Access
```

```python
arr[:, 4] #Column Access
```

Slicing of Matrix

```python
arr[0:2, 1:3]
```

```python
arr[2:4, 3:5]
```

```python
arr[1:, :]
```

```python
arr[:, :2]
```

```python
arr[::2, ::2]
```
