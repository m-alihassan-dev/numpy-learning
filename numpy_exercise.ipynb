```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],  # Student 1
    [19, 92, 88],  # Student 2
    [17, 76, 95],  # Student 3
    [18, 65, 70],  # Student 4
    [20, 90, 85]   # Student 5
])
```

Get The Shape of the Matrix

```python
data.shape
```

Find the Average of Age of Students


```python
np.mean(data[:, 0])
```

Extract Math Marks of All Students

```python
data[:, 1]
```

Find the Highest Science Marks


```python
np.max(data[:, 2])
```

Get Details of Student who Scored more than 90 Marks in Math


```python
data[data[:, 1] > 90]
```

Increase Math Marks of all Students by 5


```python
data[:, 1] += 5
data
```

Find How many Students are Younger than 19


```python
len(data[data[:, 0] < 19])
```

Calculate Average Marks in each Subject

```python
print("Average Math Marks:", np.mean(data[:, 1]))
print("Average Science Marks:", np.mean(data[:, 2]))
```

Get the Data of Students who Scored atleast 80 in both Subjects


```python
data[(data[:, 1] >= 80) & (data[:, 2] >= 80)]
```

Replace all Science Marks < 75 with 0

```python
data[:, 2][data[:, 2] < 75] = 0
data
```
