```python
import numpy as np
```

Stacking Arrays

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

np.stack((a, b))
```

```python
np.vstack((a, b))
```

```python
np.hstack((a, b))
```

```python
np.column_stack((a,b))
```

Splitting Arrays

```python
a = np.arange(1, 31).reshape(5, 6)
print(a)
```

```python
np.split(a, 5)
```

```python
np.array_split(a, 3)
```

```python
np.vsplit(a, 5)
```

```python
np.hsplit(a, 6)
```
