Creating Arrays from Python Lists

```python
import numpy as np

np.array([1, 2, 3, 4, 5])
```

Generating Arrays with `arange`

```python
np.arange(1, 5)
```

Initializing Arrays with Placeholder Values

```python
np.zeros(5)
```

```python
np.zeros((5, 5))
```

```python
np.ones(5)
```

```python
np.ones((2, 2))
```

Generating Random Data

```python
a = np.random.rand(5)
print(a)
```

```python
b = np.random.randn(5)
print(b)
```

```python
c = np.random.randint(1, 11, 5)
print(c)
```

```python
d = np.random.randint(1, 11, (5,10))
print(d)
```
